```python
import os
import shutil

root_dir = r"C:\Users\Charlotte\Desktop\dissertation\US_new\Aixplorer_Series_FirstFilter"

output_dir = r"C:\Users\Charlotte\Desktop\dissertation\US_new\All_Images"
os.makedirs(output_dir, exist_ok=True)

valid_ext = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')

for patient_folder in os.listdir(root_dir):
    patient_path = os.path.join(root_dir, patient_folder)
    if not os.path.isdir(patient_path):
        continue

    for subfolder, _, files in os.walk(patient_path):
        if output_dir in subfolder:
            continue

        folder_name = os.path.basename(subfolder)
        for file in files:
            if file.lower().endswith(valid_ext):
                src = os.path.join(subfolder, file)

                new_name = f"{patient_folder}_{folder_name}_{file}"
                dst = os.path.join(output_dir, new_name)

                base, ext = os.path.splitext(dst)
                counter = 1
                while os.path.exists(dst):
                    dst = f"{base}_{counter}{ext}"
                    counter += 1

                shutil.copy2(src, dst)

print("All images have been successfully renamed and concentrated into the All_Images folder!")
```
