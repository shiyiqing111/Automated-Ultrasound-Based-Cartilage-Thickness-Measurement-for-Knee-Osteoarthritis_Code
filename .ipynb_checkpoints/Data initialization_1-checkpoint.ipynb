```python
import os
import pydicom
import numpy as np
from PIL import Image

root_dir = "C:/Users/Charlotte/Desktop/dissertation/US_new/Aixplorer_Series"

def is_dicom_file(filepath):
    try:
        with open(filepath, 'rb') as f:
            return f.read(132)[128:132] == b'DICM'
    except:
        return False

def dcm_to_jpg(dcm_path, jpg_path):
    try:
        ds = pydicom.dcmread(dcm_path, force=True)
        arr = ds.pixel_array.astype(float)
        scaled = (np.maximum(arr, 0) / arr.max()) * 255.0
        img = Image.fromarray(np.uint8(scaled))
        img.save(jpg_path, "JPEG")
        return True
    except Exception as e:
        print(f"Fail: {dcm_path} -> {e}")
        return False

for root, dirs, files in os.walk(root_dir):
    for file in files:
        file_path = os.path.join(root, file)
        if is_dicom_file(file_path):
            jpg_path = os.path.splitext(file_path)[0] + ".jpg"
            if dcm_to_jpg(file_path, jpg_path):
                os.remove(file_path)
                print(f"Converted and deleted: {file_path}")
            else:
                print(f"Skip unsuccessful files: {file_path}")

print("All DICOM files have been converted to JPG")
```
